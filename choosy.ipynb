```python
from model import Fitness

# Data visualization tools.
import seaborn as sns
import seaborn.objects as so
import numpy as np
import pandas as pd
import janitor
import matplotlib.pyplot as plt

import altair as alt
alt.data_transformers.enable("vegafusion")
alt.renderers.enable("jupyter", offline=True)
```

```python
model = Fitness(200, 20, 20, abundance=0.25, mutation=False, choosy=True)

for _ in range(10000):
    model.step()

model_data = model.datacollector.get_model_vars_dataframe()
model_agent_data = model.datacollector.get_agent_vars_dataframe()
```

```python
agent_pops = model_data.drop(columns=['Avg Fitness', 'Population']).reset_index().rename(columns={'index': 'step'})

agent_pops.head()
```

```python
agent_pops_long = agent_pops.pivot_longer(index='step',
                                          names_to=('role', 'strategy', 'stat'),
                                          names_sep='_')

agent_pops_long
```

```python
alt.Chart(agent_pops_long.loc[agent_pops_long['stat'] == 'pop']).mark_line(
).encode(
    alt.X('step:T'),
    alt.Y('value:Q'),
    color='role:N',
    strokeDash='strategy:N'
).properties(
    width=800
)
```
